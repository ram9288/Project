```python
bins =[0,100000,300000,float('inf')]
labels = ['low', 'Med', 'High']
df['Income band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels , right=False)
```

```python
df['Income band'].value_counts().plot(kind='bar')
```

```python
# Examine the distribution of unique cataegories in categorical colums
categorical_cols= df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income band"]].columns
for col in categorical_cols:
  print(f"Value Count for'{col}':")
  display(df[col].value_counts())
```

```python
# Histogram of value count for different Occupation

for col in categorical_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(8,4))
  df[col].value_counts().plot(kind='bar')
  plt.title(f'Histogram of {col} Count')
  plt.xlabel(col)
  plt.ylabel("count")
  plt.show()
```

Numerical Analysis

```python
numerical_cols= ['Estimated Income','Superannuation Savings','Credit Card Balance','Saving Accounts','Foreign Currency Account','Business Lending']
# Univariate analysis and visualizaton
plt.figure(figsize=(20,10))
for i, col in enumerate(numerical_cols):
  plt.subplot(2,3,i+1)
  sns.histplot(df[col], kde=True)
  plt.title(col)
plt.tight_layout()
plt.show()
```

Heatmaps

```python
numerical_cols= ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans', 'Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
numerical_cols= ['Estimated Income','Superannuation Savings','Credit Card Balance','Saving Accounts','Foreign Currency Account','Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
